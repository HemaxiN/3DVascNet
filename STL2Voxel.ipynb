```
#000 012
import cv2
path = 'output_0'
from scipy.ndimage.interpolation import rotate

import numpy as np
a = []
for im in range(0,10):
    image = cv2.imread(path+str(im)+'.png')
    #plt.imshow(a[14])
    image = cv2.flip(image,1)
    image = rotate(image, 90, mode='constant', reshape=True)
    print(path+str(im))
    a.append(image[:,:,0])

path = 'output_'
for im in range(10,99):
    image = cv2.imread(path+str(im)+'.png')
    image = cv2.flip(image,1)
    image = rotate(image, 90, mode='constant', reshape=True)

    print(path+str(im))
    a.append(image[:,:,0])

# path = 'WT_retina_5/output_'
# for im in range(100, 103):
#   image = cv2.imread(path+str(im)+'.png')
#   print(path+str(im))
#   a.append(image[:,:,0])
```

```
a = np.asarray(a)
a = a.transpose(1,2,0)
np.shape(a)
```

```
a = a.astype('uint8')
from tifffile import imwrite
imwrite('mask_WT_AngII_ret1.tif', a)  #write the 3D image as a .tif file
```

```
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (20,20)
image = np.max(a, axis=2)  "2D projection of the 3D mask"
plt.imshow(image, cmap='gray')
```
